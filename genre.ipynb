```python
import os
import librosa
import soundfile as sf
import subprocess
import pandas as pd
import utils


AUDIO_DIR = r'fma_medium path'
tracks = utils.load(r'tracks.csv path')


os.makedirs('Samples', exist_ok=True)

# small = tracks['set', 'subset'] == 'small'
medium = tracks['set', 'subset'] == 'medium'
# large = tracks['set', 'subset'] == 'large'


y_medium = tracks.loc[medium, ('track', 'genre_top')]


sr = 44100
for track_id, genre in y_medium.items():
    if pd.isna(genre):  
        continue

 
    safe_genre = genre.replace('/', ' or ')  


    genre_dir = os.path.join('Samples', safe_genre)
    os.makedirs(genre_dir, exist_ok=True)

    
    mp3_filename = utils.get_audio_path(AUDIO_DIR, track_id)
    out_wav_filename = os.path.join(genre_dir, f'{track_id}.wav')

   
    cmd = ['ffmpeg', '-hide_banner', '-loglevel', 'panic', '-i', mp3_filename, out_wav_filename]
    try:
        subprocess.run(cmd, check=True)
        print(f"Converted {mp3_filename} to {out_wav_filename}")
    except subprocess.CalledProcessError as e:
        print(f"FFmpeg failed for {mp3_filename}, skipped. Error: {e}")
        continue

   
    try:
        data, _ = librosa.load(out_wav_filename, sr=sr, mono=True)
        sf.write(out_wav_filename, data, sr)
    except Exception as e:
        print(f"Error processing {out_wav_filename}: {e}")
        continue
```
